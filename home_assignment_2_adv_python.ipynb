### Task 3.1

````python
x_column = 'volatile acidity'
y1_column = 'citric acid'
y2_column = 'residual sugar'

# Create a scatter plot
plt.scatter(loaded_df[x_column], loaded_df[y1_column], label=f'{y1_column}', alpha=0.5, color='blue')
plt.scatter(loaded_df[x_column], loaded_df[y2_column], label=f'{y2_column}', alpha=0.5, color='orange')

# Add title and labels
plt.title(f'Scatter Plot of {y1_column} and {y2_column} against {x_column}')
plt.xlabel(x_column)
plt.ylabel('Y Coordinates')
plt.legend()

# Show the plot
plt.show()
````

### Task 3.2

Choose 2 columns from the dataset. 

On the **same figure** draw 3 plots:
- histogram of the first column
- histogram of the second column
- a scatter plot where one column is the X axis coordinates and the other column is the Y axis coordinates.

**Make sure to add a title, a legend and axis labels to all your plots**

#### HINT
- See lecture 07 again.
- Use: `fig, ax = plt.subplots(3,1)`

````python
selected_columns = ['volatile acidity', 'citric acid']

# Create subplots
fig, ax = plt.subplots(3, 1, figsize=(8, 12))

# Plot histograms
for i, col in enumerate(selected_columns):
    ax[i].hist(loaded_df[col], bins=20, color='blue', alpha = 0.5)
    ax[i].set_title(f'Histogram of {col}')
    ax[i].set_xlabel(col)
    ax[i].set_ylabel('Frequency')

# Plot scatter plot
ax[2].scatter(loaded_df[selected_columns[0]], loaded_df[selected_columns[1]], alpha=0.5, label='Scatter Plot')
ax[2].set_title(f'Scatter Plot between {selected_columns[0]} and {selected_columns[1]}')
ax[2].set_xlabel(selected_columns[0])
ax[2].set_ylabel(selected_columns[1])
ax[2].legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show() 
````

## Part 4 - Visualization - Correlation Heatmap

### Task 4.1

Create a correlation dataframe for the dataset (all columns against all columns).

Use seaborn to plot the correlation heatmap.

#### Hint
We did this in class, see lecture 7, "HeatMap" section.

````python
matrix = loaded_df.corr()

# Plot a heatmap using Seaborn
plt.figure(figsize=(14,12 ))
sns.heatmap(matrix, annot=True, cmap='magma',  vmin=0, vmax=1)

# Add a title
plt.title("Correlation Heatmap")

# Show the plot
plt.show()
````
